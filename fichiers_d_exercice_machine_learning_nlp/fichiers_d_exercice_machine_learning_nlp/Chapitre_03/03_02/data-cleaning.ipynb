```python
import pandas as pd
pd.set_option('display.max_colwidth', 100)

data = pd.read_csv("../../Data/SMSSpamCollection.txt", sep='\t', header=None)
data.columns = ['label', 'Content']

data.head()
```

```python
import string
string.punctuation
```

<h4>Step 1: First , im going to clean the data "Content variable" by deleting all of character that are including to the string_punctuation</h4>

```python
# creating the function to remove all oh string_punctuation character
def Remove_punctuation(text):
    result = "".join([ch for ch in text if ch not in string.punctuation])
    return result

data['Content_clean']=data['Content'].apply(lambda x:Remove_punctuation(x))
```

```python
data
```

<h4>Step2: Removing all STOPWORDS</h4>

```python
import re
def toknizer(text):
    words = re.split('\W+',text)
    return words

data['Content_tokenized'] = data['Content_clean'].apply(lambda x: toknizer(x))
```

```python
data.head()
```

<h4>Step3: We are going to remove English_StopWords from the content_tokenized "LIST"</h4>


```python
import nltk
```

```python
en_stopwords = nltk.corpus.stopwords.words('english')
```

```python
def Remove_StopWords(list_text):
    result = [word for word in list_text if word not in en_stopwords]
    return result


data['Content_No_StopWords'] = data['Content_tokenized'].apply(lambda x: Remove_StopWords(x))
data
```

<h4>Step 4: We are going to use steamming method</h4>


```python
# I will merge All of the functions I created above in only one function
def clean_email(email):
    #first: Removing all of strin.punctuation
    content_clean = "".join([word for word in email if word not in string.punctuation])
    #Second: Spleting the result into one list witch contains words like elements
    content_tokenizes = re.split('\W+',content_clean)
    #Third: Removing StopWords
    text = [word for word in content_tokenizes if word not in en_stopwords]
    return text
```

```python
data2 = pd.read_csv("../../Data/SMSSpamCollection.txt", sep='\t', header=None)
data2.columns = ['label', 'Content']
data2
```

```python
data2['Content_Non_StopWords'] = data2['Content'].apply(lambda x: clean_email(x)) 
```

```python
data2
```

```python
ps = nltk.PorterStemmer()
def steamming(words):
    result = [ps.stem(word) for word in words]
    return result

data2['Content_Stemmed'] = data2['Content_Non_StopWords'].apply(lambda x: steamming(x))
data2
```
